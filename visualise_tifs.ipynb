```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.plot import show
import glob
```

```python
def visualise_plots(tif_files, titles, savefig_path):

    # Create a figure with 4 rows and 2 columns
    fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 16))

    # Flatten the axes array to easily access subplots in a loop
    axes = axes.flatten()

    # Loop through the files and plot each on a subplot
    for i, (tif_file, ax, title) in enumerate(zip(tif_files, axes, titles)):
        with rasterio.open(tif_file) as src:
            image = src.read(1)  # Read the first band
            show(image, ax=ax, cmap='gray')  # Show the image on the subplot
            ax.set_title(title)  # Set the title for each subplot
            ax.axis('off')  # Turn off the axis for a cleaner look

    # Adjust layout
    plt.tight_layout()

    # Save the figure
    plt.savefig(savefig_path+'.png', dpi=300, bbox_inches='tight')

    # Show the figure
    # plt.show()
```
